Utilizar la clase RadiusNeighborsClassifier para crear un clasificador que permita identificar la especie de Iris del conjunto de datos Iris en base al largo y ancho de los s√©palos.

Probar el funcionamiento con los radios de 0,3, 0,5, 0,7, 0,9 y 1,1.

```python
from sklearn.neighbors import RadiusNeighborsClassifier
from sklearn.datasets import load_iris
from matplotlib.colors import ListedColormap

iris_data = load_iris()
# print(iris_data.feature_names)
# print(iris_data.data)
# print(iris_data.target_names)
# print(iris_data.target)
# Solo utilizamos las columnas de largo y ancho de los sepalos
X = iris_data.data[:, 0:2]
# print(X)
y = iris_data.target
print(y)
# Graficamos para ver como estan distribuidos los datos
mapa_color = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
scatter(X[:, 0], X[:, 1], c=y, cmap = mapa_color)
```

```python
def clasificadorRadius(X, y, r):
    
    knn = RadiusNeighborsClassifier(r, outlier_label = 3)
    knn.fit(X, y)

    clasificacion = knn.predict(X)
    mapa_color = ListedColormap(['#FFFF00', '#11FF00', '#1100FF'])
    scatter(X[:, 0], X[:, 1], c=clasificacion, cmap = mapa_color)
```

```python
clasificadorRadius(X, y, 0.3)
```

```python
clasificadorRadius(X, y, 0.5)
```

```python
clasificadorRadius(X, y, 0.7)
```

```python
clasificadorRadius(X, y, 0.9)
```

```python
clasificadorRadius(X, y, 1.1)
```
